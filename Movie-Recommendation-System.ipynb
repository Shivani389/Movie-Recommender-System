```python
movies=movies[['id_x' ,'title' ,'overview_x' , 'genres_x','keywords_x']]
```

```python
import ast
ast.literal_eval
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres_x']=movies['genres_x'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords_x']=movies['keywords_x'].apply(convert)
```

```python
movies.head()
```

```python
movies['overview_x']=movies['overview_x'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres_x']=movies['genres_x'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords_x']=movies['keywords_x'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview_x']+movies['genres_x']+movies['keywords_x']
```

```python
movies.head()
```

```python
new_data=movies[['id_x','title','tags']]
```

```python
new_data
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:" ".join(x))
```

```python
new_data.head()
```

```python
new_data['tags'][0]
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:x.lower())
```

```python
new_data.head()
```

```python
new_data['tags'][0]
```

```python
new_data['tags'][1]
```

```python
#stemming
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_data['tags']=new_data['tags'].apply(stem)
```

```python
new_data['tags'][0]
```

```python
#Text Vectorization - Bag of words technique  
```

```python
#stopword removal
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_data['tags']).toarray()
```

```python
#Calculating the distance between each movies using cosine distance
#More the angle more will be the dissimilarity b/w the movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=new_data[new_data['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
